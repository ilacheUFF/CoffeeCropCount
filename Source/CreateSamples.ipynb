```python
#%pylab inline

from PIL import Image
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage.feature import match_template

#ImagenTotal = np.asarray(Image.open('MosaicoPouco.png'))  #Nome da imagem onde estão os templates
ImagenTotal = np.asarray(Image.open('MetadeAreaTotal.png'))
print(ImagenTotal.shape)


%matplotlib qt
```

# Código para gravar os templates de cafe

```python
figOriginal, ax = plt.subplots()
plt.imshow(ImagenTotal)
clicks=0

def onclick(event):
    global clicks
    print('%s click: button=%d, x=%d, y=%d, xdata=%f, ydata=%f' %
          ('double' if event.dblclick else 'single', event.button,
           event.x, event.y, event.xdata, event.ydata))
    clicks=clicks+1
    SaveTemplate(np.round(event.xdata).astype(int),np.round(event.ydata).astype(int),clicks)
    print(clicks)
    
def SaveTemplate(x,y,clicks):
    global ImagenTotal
    yis=y    
    xis=x
    print('El valor de y = ',y,'; El valor de x=',x)
    Lado=150  #Lado / 2 do quadrado onde se gera o template
    fig, ax = plt.subplots()
    plt.imshow(ImagenTotal[yis-Lado:yis+Lado,xis-Lado:xis+Lado,:])
    Nome='Cafe_10'
    Nome=Nome+str(clicks)+'.png'
    ax.axis('off')
    plt.show()
    fig.savefig(Nome, bbox_inches='tight')   # save the figure to file
    plt.close(fig) 

cid = figOriginal.canvas.mpl_connect('button_press_event', onclick)
```

# Segmentar las imagenes

```python
def SaveTemplate2(x,y,clicks):
    global ImagenTotal
    yis=y    
    xis=x
    print('El valor de y = ',y,'; El valor de x=',x)
    Lado=150  #Lado / 2 do quadrado onde se gera o template
    fig, ax = plt.subplots()
    plt.imshow(ImagenTotal[yis-Lado:yis+Lado,xis-Lado:xis+Lado,:])
    Nome='Total_00'
    Nome=Nome+str(clicks)+'.png'
    ax.axis('off')
    plt.show()
    fig.savefig(Nome, bbox_inches='tight')   # save the figure to file
    plt.close(fig)
clicks=0
for x in range(150,150*7,75):
    print ('*****')
    for y in range(150,150*6,75):
        SaveTemplate2(x,y,clicks)
        clicks=clicks+1
```
